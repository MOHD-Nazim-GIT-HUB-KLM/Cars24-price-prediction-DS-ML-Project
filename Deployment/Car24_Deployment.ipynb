```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    print(f"{column}: {len(outliers)} outliers")
    return outliers

# Check for outliers in selected columns
for col in ['price', 'kilometerdriven', 'discountprice']:
    detect_outliers_iqr(df, col)
```

```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    cleaned_data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    print(f"{column}: Removed {len(data) - len(cleaned_data)} outliers")
    return cleaned_data


df = remove_outliers_iqr(df, 'price')
df = remove_outliers_iqr(df, 'kilometerdriven')
df = remove_outliers_iqr(df, 'discountprice')
```

```python
import pickle
from fastapi import FastAPI
from pydantic import BaseModel
import pandas as pd
```

```python

# Load the trained model
with open("car_price_model.pkl", "rb") as f:
    model = pickle.load(f)

app = FastAPI(title="Car Price Prediction API")
```

```python



# Define request schema according to your features
class CarFeatures(BaseModel):
    make: int
    model: int
    city: int
    year: int
    fueltype: int
    kilometerdriven: int
    ownernumber: int
    transmission: int
    bodytype: int
    isc24assured: int
    benefits: int
    discountprice: int
    car_age: int
```

```python


@app.post("/predict")
def predict(features: CarFeatures):
    # Convert to DataFrame
    data = pd.DataFrame([features.dict()])
    prediction = model.predict(data)[0]
    return {"predicted_price": prediction}
```
